```python
from dotenv import load_dotenv
from elasticsearch import Elasticsearch
import requests, os
import pandas as pd
```

```python
# You need to add the infos in the .env file
load_dotenv()

ES_HOST = os.getenv("ES_HOST")
ES_PASSWORD = os.getenv("ES_PASSWORD")
ES_USER = os.getenv("ES_USER")
es = Elasticsearch(ES_HOST, http_auth=(ES_USER, ES_PASSWORD))
INDEX_PUBLICATIONS = f'scanr-publications'

from elasticsearch import Elasticsearch, helpers
es = Elasticsearch(ES_HOST, http_auth=(ES_USER, ES_PASSWORD))
```

```python
INDEX_PUBLICATIONS
```

```python
# Example of a query with an RNSR


body={'size': 1000,
 'query': {'bool': {
     'filter': [
     
          {"term": {"affiliations.id.keyword":"200320607R"}},
        {'range': {'year': {'gte': 2020, 'lte': 2024}}}
         
     ]
 }},
     }

res = es.search(index=INDEX_PUBLICATIONS, request_cache=False, request_timeout=6000, body=body)
```

```python
res['hits']['total']
```

```python
all_data =  [k['_source'] for k in res['hits']['hits']]
df = pd.DataFrame(all_data)
```

```python
df_simple = df[['id', 'title', 'year', 'projects', 'domains', 'authors', 'acknowledgments', 'topics', 'cited_by_counts_by_year']]
```
